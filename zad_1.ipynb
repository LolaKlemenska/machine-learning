```python
import pandas as pd
import numpy as np
```

```python
dane = pd.read_excel(r'/home/lolusia/uczenie_maszynowe/Książka 2.xlsx')
dane
```

```python
dane['obserwacje'][9] = 3.32
```

```python
dane['entalpia'] = (dane['entalpia'] - dane['entalpia'].mean()) / dane['entalpia'].std()
dane['elektroujemnosc'] = (dane['elektroujemnosc'] - dane['elektroujemnosc'].mean()) / dane['elektroujemnosc'].std()
dane
```

```python
predicted = pd.Series(2.466 + 0.244 *dane['entalpia'][i] + 0.394*dane['elektroujemnosc'][i] for i in range(dane.shape[0]))
predicted
```

```python
tabela = pd.concat([dane, predicted], axis=1)
tabela.columns = ['nazwa', 'entalpia', 'elektroujemnosc', 'obserwacje', 'obliczone']
tabela
```

```python
R2 = 1 - (((tabela['obliczone']-tabela['obserwacje'])**2).sum()/((tabela['obserwacje']-tabela['obserwacje'].mean())**2).sum())
RMSEc = np.sqrt(((tabela['obliczone']-tabela['obserwacje'])**2).sum()/tabela.shape[0])
Sm2 = ((tabela['obliczone']-tabela['obliczone'].mean())**2).sum()/(tabela.shape[0]-1)
Se2 = ((tabela['obserwacje']-tabela['obliczone'])**2).sum()/(tabela.shape[0]-2-1)
F = Sm2/Se2
```

```python
print(f'Współczynnik determinacji dla modelu wynosi {R2},\nśredni kwadratowy błąd kalibracji wynosi {RMSEc},\na istotność statystyczna modelu według testu F-Snedecora wynosi {F}')
```
